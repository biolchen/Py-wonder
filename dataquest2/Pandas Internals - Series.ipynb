```python
import pandas as pd
```

```python
fandango = pd.read_csv('fandango_score_comparison.csv')
fandango.head()
```

```python
series_film = fandango['FILM']
series_rt = fandango['RottenTomatoes']
print(series_film, series_rt)
```

```python
series_custom = pd.Series(series_rt.values, index=series_film.values)
print(series_custom)
```

```python
fiveten = series_custom[5:10]
print(fiveten)
```

```python
sorted_by_index = series_custom.reindex(sorted(series_custom.index.tolist()))
print(sorted_by_index)
```

```python
sc2 = series_custom.sort_index()
sc3 = series_custom.sort_values()
print(sc2[:10], sc3[:10])
```

```python
series_normalized = series_custom / 100 * 5
print(series_normalized)
```

```python
rt_critics = pd.Series(fandango['RottenTomatoes'].values, index=fandango['FILM'])
rt_users = pd.Series(fandango['RottenTomatoes_User'].values, index=fandango['FILM'])
rt_mean = (rt_critics + rt_users) / 2
print(rt_mean)
```
